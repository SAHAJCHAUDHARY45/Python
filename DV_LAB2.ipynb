Task 1: Write a python code using matplotlib mentioning four-line charts like dashed, dotted, dotted
dash and simple line.

```python
import matplotlib.pyplot as plt  # Import matplotlib for plotting

x = [1, 2, 3, 4, 5]  # Define X-axis values
y1 = [10, 20, 30, 40, 50]  # Y-values for first line
y2 = [5, 15, 25, 35, 45]  # Y-values for second line
y3 = [3, 13, 23, 33, 43]  # Y-values for third line
y4 = [7, 17, 27, 37, 47]  # Y-values for fourth line

plt.plot(x, y1, linestyle='-', color='b', label='Solid Line')  # Solid line
plt.plot(x, y2, linestyle='--', color='r', label='Dashed Line')  # Dashed line
plt.plot(x, y3, linestyle=':', color='g', label='Dotted Line')  # Dotted line
plt.plot(x, y4, linestyle='-.', color='purple', label='Dash-dot Line')  # Dash-dot line

plt.xlabel("X-axis")  # Label for X-axis
plt.ylabel("Y-axis")  # Label for Y-axis
plt.title("Four Different Line Styles")  # Title of the chart
plt.legend()  # Show the legend
plt.show()  # Display the plot
```

Task 2: Write code for line chart to generate 70 numbers with X- axis and Y-axis is mx+c, were m=2
and c=0.3, Draw the four different line charts using matplotlib library. (Take Y1=X, Y2=X2, Y3=X3 and
Y4=sqrt(X)).

```python
import numpy as np

x = np.arange(1, 71)  # Generate 70 numbers from 1 to 70
y1 = x  # y = x
y2 = x**2  # y = x^2
y3 = x**3  # y = x^3
y4 = np.sqrt(x)  # y = sqrt(x)

plt.plot(x, y1, label="y = x")  # Plot line y = x
plt.plot(x, y2, label="y = x^2")  # Plot line y = x^2
plt.plot(x, y3, label="y = x^3")  # Plot line y = x^3
plt.plot(x, y4, label="y = sqrt(x)")  # Plot line y = sqrt(x)

plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.legend()
plt.title("Different Mathematical Functions")
plt.show()
```

Task 3: Draw the following house using matplotlib line charts: (Note You can take any another house
also with one window to draw.)


```python
import matplotlib.pyplot as plt

# Define house structure coordinates (walls)
house_x = [1, 1, 5, 5, 1]
house_y = [1, 5, 5, 1, 1]

# Define roof coordinates (triangle)
roof_x = [1, 3, 5]
roof_y = [5, 7, 5]

# Define door coordinates (Increased height)
door_x = [2, 2, 3.5, 3.5, 2]
door_y = [1, 4, 4, 1, 1]  # Door height increased from 3 to 4

# Define window coordinates (Square shape)
window_x = [4, 4, 4.8, 4.8, 4]
window_y = [3, 4, 4, 3, 3]

# Plot the house structure
plt.plot(house_x, house_y, color='brown', linewidth=2, label="House Walls")
plt.plot(roof_x, roof_y, color='red', linewidth=2, label="Roof")
plt.plot(door_x, door_y, color='black', linewidth=2, label="Taller Door")  # Updated door
plt.plot(window_x, window_y, color='blue', linewidth=2, label="Window")

# Set labels and title
plt.title("House with Taller Door and Window Using Matplotlib")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Display legend and grid
plt.legend()
plt.grid()

# Show the plot
plt.show()
```

1. How do you add multiple lines to a single chart?

   Call the plot() function multiple times, each with different sets of x and y
  data.

   Each plot() call will add a new line to the chart.

   Different lines can be distinguished by using different colors, styles, or
  labels.
2. How can you add labels, legends, and titles to a line chart?

  Labels:

   Use xlabel() to label the X-axis.
   Use ylabel() to label the Y-axis.

  Legend:

   Use legend() to add a legend, associating labels with specific lines.

   Each line can be labeled using the label argument in plot().

  Title:

   Use title() to add a title to the chart, giving context to what the chart
  represents.

3. How can you customize the appearance of the line (color,
style, markers)?

  Color:

   Use color to set the line color (e.g., 'blue', 'red', or '#FF5733').

  Line Style:

   Use linestyle to set the line style:

  o Solid line ('-')

  o Dashed line ('--')

  o Dotted line (':')

  o Dash-dot line ('-.')

  Markers:

   Use marker to add markers to the line (e.g., 'o', 's', '^').
  
   Use markersize to adjust the size of the markers.
