```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)  # Generates 100 values between 0 and 10
y = np.sin(x)  # Computes sine values for the generated x values

plt.plot(x, y, label="sin(x)")  # Plots x against y with a label
plt.xlabel("X-axis")  # Labels the x-axis
plt.ylabel("Y-axis")  # Labels the y-axis
plt.title("Line Plot")  # Sets the title of the plot
plt.legend()  # Displays the legend
plt.show()  # Shows the plot
```

```python
x = np.random.rand(50)  # Generates 50 random x values between 0 and 1
y = np.random.rand(50)  # Generates 50 random y values between 0 and 1

plt.scatter(x, y, color="red", marker="o")  # Plots the scatter points
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter Plot")
plt.show()
```

```python
categories = ['A', 'B', 'C', 'D']  # Category labels
values = [10, 20, 15, 25]  # Corresponding values

plt.bar(categories, values, color="blue")  # Creates a bar chart
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Bar Plot")
plt.show()
```

```python
data = np.random.randn(1000)  # Generates 1000 random numbers with normal distribution

plt.hist(data, bins=30, color="green", edgecolor="black")  # Creates a histogram
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram")
plt.show()
```

```python
labels = ['Apple', 'Banana', 'Cherry', 'Date']  # Labels for pie slices
sizes = [30, 20, 40, 10]  # Size of each slice

plt.pie(sizes, labels=labels, autopct="%1.1f%%", colors=["red", "yellow", "pink", "brown"])
plt.title("Pie Chart")
plt.show()
```

```python
data = [np.random.randn(100) for _ in range(4)]  # Generates 4 groups of 100 random numbers

plt.boxplot(data)  # Creates a box plot
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Box Plot")
plt.show()
```

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.fill_between(x, y1, y2, color="lightblue", alpha=0.5)  # Fills the area between two curves
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Area Plot")
plt.show()
```

```python
data = np.random.rand(10, 10)  # Creates a 10x10 matrix with random values

plt.imshow(data, cmap="coolwarm", interpolation="nearest")  # Displays the matrix as an image
plt.colorbar()  # Adds a color scale bar
plt.title("Heatmap")
plt.show()
```

```python
x = np.arange(10)  # x values from 0 to 9
y = np.random.randint(1, 10, size=10)  # Random y values

plt.stem(x, y, linefmt="r-", markerfmt="ro", basefmt="k-")  # Creates a stem plot
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Stem Plot")
plt.show()
```

```python
theta = np.linspace(0, 2*np.pi, 100)  # Angle values
r = np.abs(np.sin(2*theta))  # Radius values

plt.polar(theta, r)  # Creates a polar plot
plt.title("Polar Plot")
plt.show()
```
