**Task 1: Line Chart with simple Annotations**

**Task 3: Line Chart for Time-Series Data. (Take Date & Value
as X & Y-axis, draw a chart for Day, Month and Year)**

```python
import pandas as pd
# Example time-series data
data = {'Date': pd.date_range(start='2025-01-01', periods=7),'Value': [10, 12, 15,18,20,22,25]}
df = pd.DataFrame(data)
# Plotting the line chart
plt.plot(df['Date'], df['Value'], marker='o', color='green')
plt.title("Time-Series Line Chart")
plt.xlabel("Date")
plt.ylabel("Value")
plt.xticks(rotation='vertical')
plt.grid()
plt.show()
```

```python
# Example time-series data
data = {
'Date': pd.date_range(start='2025-01-01', periods=10,freq='D'),'Value': [10, 12,15, 18, 20, 21, 25, 28, 30, 32]}
df = pd.DataFrame(data)
# Plotting the time-series line chart
plt.figure(figsize=(6, 4))
plt.plot(df['Date'], df['Value'], marker='o', color='blue', label='Value Over Time')
# Adding titles and labels
plt.title("Time-Series Line Chart", fontsize=16)
plt.xlabel("Date", fontsize=12)
plt.ylabel("Value", fontsize=12)
# Rotating the x-axis labels for better readability
plt.xticks(rotation='vertical')
plt.grid(alpha=0.5)
plt.legend()
# Display the plot
plt.tight_layout()
plt.show()
```

**Task 4: Draw line chart with y = np.sin(x). Fill the pink color
in waves.**

```python
import numpy as np
# Data
x = np.linspace(0, 10, 100)
y = np.sin(x)
# Create line chart with shading
plt.plot(x, y, label='Sine Wave', color='blue')
plt.fill_between(x, y, color='red', alpha=0.2)
plt.title("Line Chart with Shaded Area")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.grid()
plt.show()
```

**Task 5: Plot sin(x) and cos(x) from 0 to 2 on the same
graph. Use fill_between() to shade the area between the two
curves where sin(x) is greater than cos(x). Add
labels and a title.
**

```python
# Define x values from 0 to 2
x = np.linspace(0, 2 * np.pi, 1000)
sin_x = np.sin(x)
cos_x = np.cos(x)
plt.figure(figsize=(8, 5))
plt.plot(x, sin_x, label='sin(x)', color='blue')
plt.plot(x, cos_x, label='cos(x)', color='red')
plt.fill_between(x, sin_x, cos_x, where=(sin_x > cos_x), color='Green', alpha=0.5)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Sine and Cosine Functions with Shaded Region Where sin(x) > cos(x)')
plt.legend()
plt.grid()
plt.show()
```

**Task 6: Draw multiple plots with three rows and one
column. Show the sine of any 30 integers in the first plot,
the cosine of t same 30 integers in the second plot, and the
tangent of the same 30 integers in the 3rd plot. Save the
subplots.**


```python
# Define 30 integer values
x = np.arange(30)
sin_x = np.sin(x)
cos_x = np.cos(x)
tan_x = np.tan(x)
# Create subplots with 3 rows and 1 column
fig,axes = plt.subplots(3, 1, figsize=(8, 10))
# First plot: Sine
axes[0].plot(x, sin_x, label='sin(x)', color='blue')
axes[0].set_title('Sine of 30 Integers')
axes[0].set_xlabel('x')
axes[0].set_ylabel('sin(x)')
axes[0].legend()
axes[0].grid()
# Second plot: Cosine
axes[1].plot(x, cos_x, label='cos(x)', color='red')
axes[1].set_title('Cosine of 30 Integers')
axes[1].set_xlabel('x')
axes[1].set_ylabel('cos(x)')
axes[1].legend()
axes[1].grid()
# Third plot: Tangent
axes[2].plot(x, tan_x, label='tan(x)', color='green')
axes[2].set_title('Tangent of 30 Integers')
axes[2].set_xlabel('x')
axes[2].set_ylabel('tan(x)')
axes[2].legend()
axes[2].grid()
plt.tight_layout()# for avoid overlapping.
plt.show()
```

**Task 7: Generate the script for plots two plots. The outer
plot displays the square root, while the inner plot displays
the cube of 20 integers between 0 and 20.
Save the subplots.
**

```python
x=np.arange(20)
Squar_root_x=np.sqrt(x)
Cube_x=x**3
fig,ax_outer=plt.subplots(figsize=(8,6))
ax_outer.plot(x,Squar_root_x,label='Square(x)',color='green')
ax_outer.set_title('Square Root Function')
ax_outer.set_xlabel('X')
ax_outer.set_ylabel('Sqrt(x)')
ax_outer.legend()
ax_outer.grid()
ax_inner = fig.add_axes([0.55, 0.2, 0.3, 0.3]) # [left, bottom, width, height]
ax_inner.plot(x,Cube_x,label='cube(x)',color='red')
ax_inner.set_title('Cube Function')
ax_inner.set_xlabel('x')
ax_inner.set_ylabel('x^3')
ax_inner.legend()
ax_inner.grid()
plt.show()
```
